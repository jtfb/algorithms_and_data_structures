```python
# Your answer here
class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right
        
def is_duplicate(root: TreeNode) -> int:
  # TODO

    # Store visited values and queue
    visited = set()
    queue = [root]

    while queue:
        # Pop the first node from the queue
        node = queue.pop(0)

        # Check if first node is visited
        if node.val in visited:
            # Return value if duplicated
            return node.val
        # Add current node to the visited set
        visited.add(node.val)

        # In case current node has a left child, add it to the queue
        if node.left:
            queue.append(node.left)
        # In case current node has a right child, add it to the queue
        if node.right:
            queue.append(node.right)
            
    # In case there is no duplicates in the tree, return -1
    return -1

# Test it
root = TreeNode(2)
root.left = TreeNode(3)
root.right = TreeNode(3)
root.left.left = TreeNode(6)
root.left.right = TreeNode(6)
root.right.left = TreeNode(7)
root.right.right = TreeNode(8)

print(is_duplicate(root))
```


-   Explain why your solution works


# Your answer here
Answering the question at the top, it's a BFS (or Level Order Traversal) as it traverses nodes present in the same level before traversing the next level. It works because I've tested here building an example from my tree. I'm using Breadth First Search (BFS) as I did test depth first search (DFS) function and it got me wrong answers when having more than one duplicate. BFS ensures that I get the duplicate value closest to the root first as BFS process nodes level by level, starting from the root. The "visited" set tracks all the values I have visited.


-   Explain the problemâ€™s time and space complexity


# Your answer here
Time and Space complexity here is linear -> O(n).
- Time Complexity: Each node is visited exactly once. The work done at each node is constant O(n). Checking and adding elements to the "visited" set is O(1)
- Space Complexity: the worst case, "visited" and "queue" would store all the values in the tree, so would be O(n)
